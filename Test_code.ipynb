```python
# Initialize an empty DataFrame
df = pd.DataFrame(columns=['id', 'type', 'sectionId', 'sectionName', 'webPublicationDate', 'webTitle', 'pillarId', 'pillarName'])
```

```python
for i in range(1, 3260):
    response = requests.get('https://content.guardianapis.com/search?page={}&q=justin%20Trudeau&api-key=test'.format(i))
    if response.status_code == 200:
        # Extract relevant data from the response
        json_data = response.json()
        if 'response' in json_data and 'results' in json_data['response']:
            temp_df = pd.DataFrame(json_data['response']['results'], columns=['id', 'type', 'sectionId', 'sectionName', 'webPublicationDate', 'webTitle', 'pillarId', 'pillarName'])
            df = pd.concat([df, temp_df], ignore_index=True)
    else:
        print("Error fetching data from page", i)
```

# Ques 7: If so, show these. Why are these unusual? (Define for yourself what you want to show by ordinary or unusual).

1. Unusual events defined by the anomaly in the number of articles published.
2. To define anomaly, here is the approach:
    a. Created a rolling window of 7 days to obtain the moving average.
    b. For any time point where the number of articles published is greater than 2 standard deviations from the moving average,
    it is defined as an anomaly.
    c. This can be further examined by changing the moving average window size.

```python
import numpy as np
# moving average with a window size of 7 days
articles_count['Moving Average'] = articles_count['No. of articles'].rolling(window=7).mean()

# the deviation from the moving average
articles_count['Deviation'] = articles_count['No. of articles'] - articles_count['Moving Average']

# threshold for anomaly detection (e.g., 2 standard deviations)
threshold = 2 * articles_count['Deviation'].std()

# anomalies
anomalies = articles_count[np.abs(articles_count['Deviation']) > threshold]

# time series with anomalies highlighted
plt.figure(figsize=(25, 8))
plt.plot(articles_count['webPublicationDate'], articles_count['No. of articles'], marker='o', linestyle='-', label='Number of Articles')
plt.plot(articles_count['webPublicationDate'], articles_count['Moving Average'], color='red', linestyle='--', label='Moving Average')
plt.scatter(anomalies['webPublicationDate'], anomalies['No. of articles'], color='red', label='Anomalies')
plt.title('Evolution of Number of Articles Over Time with Anomalies (Using Moving Average)')
plt.xlabel('Date')
plt.ylabel('Number of Articles')
plt.grid(True)
plt.xticks(rotation=45)
plt.legend()
plt.tight_layout()
plt.show()
```

In the above diagram we have considered all the points where the value is either > +2 SD from mean(moving average 7 days) or < -2 SD as anomaly.

Here is the assumption for a person if less number of articles (lesser than 2 sd from mean) published then it will not be anomaly as on that specific date he might not have done any significant event or given any significant statement.

<b> Hence we will only look at cases where the value of number of articles published is greater than 2 sd </b>

```python
import numpy as np
# Calculate the moving average with a window size of 7 days
articles_count['Moving Average'] = articles_count['No. of articles'].rolling(window=7).mean()

# Calculate the deviation from the moving average
articles_count['Deviation'] = articles_count['No. of articles'] - articles_count['Moving Average']

# Define a threshold for anomaly detection (e.g., 2 standard deviations)
threshold = 2 * articles_count['Deviation'].std()

# Identify anomalies
anomalies_updated = articles_count[articles_count['Deviation']> threshold]
```

```python
anomalies_updated["webPublicationDate"] = pd.to_datetime(anomalies_updated["webPublicationDate"])

anomalies_updated[anomalies_updated["webPublicationDate"] =="2020-03-15"]
```

```python
import plotly.graph_objects as go

# Create Plotly figure
fig = go.Figure()

# Add trace for the number of articles
fig.add_trace(go.Scatter(x=articles_count['webPublicationDate'],
                         y=articles_count['No. of articles'],
                         mode='lines+markers',
                         name='Number of Articles'))

# Add trace for moving average
fig.add_trace(go.Scatter(x=articles_count['webPublicationDate'],
                         y=articles_count['Moving Average'],
                         mode='lines',
                         line=dict(color='red', dash='dash'),
                         name='Moving Average'))

# Filter anomalies where the data is greater than +2 standard deviations
anomalies_greater_than_2_std = anomalies[anomalies['Deviation'] > threshold]

# Add scatter trace for anomalies greater than +2 standard deviations
fig.add_trace(go.Scatter(x=anomalies_greater_than_2_std['webPublicationDate'],
                         y=anomalies_greater_than_2_std['No. of articles'],
                         mode='markers',
                         marker=dict(color='red', size=10),
                         name='Anomalies > +2 Std'))

# Update layout
fig.update_layout(title='Evolution of Number of Articles Over Time with Anomalies > +2 Std (Using Moving Average)',
                  xaxis_title='Date',
                  yaxis_title='Number of Articles',
                  xaxis=dict(tickangle=-45),
                  hovermode='x',
                  template='plotly_white')

# Show the Plotly figure
fig.show()
```
